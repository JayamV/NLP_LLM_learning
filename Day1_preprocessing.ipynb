```python
def myLower(mypara):
  res=""
  for letter in mypara:
    if ('A' <=letter) and (letter <= 'Z'):
      res+=chr(ord(letter)+32)
    else:
      res+=letter

  return res


#variation using mapping
def to_lowercase(s):
    upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
    lower = "abcdefghijklmnopqrstuvwxyz"

    result = ""
    for ch in s:
        if ch in upper:
            result += lower[upper.index(ch)]
        else:
            result += ch
    return result

print(to_lowercase("HeLLo WORLD"))

#using dict
upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
lower = "abcdefghijklmnopqrstuvwxyz"
mapping = dict(zip(upper, lower))

def to_lowercase(s):
    return "".join(mapping.get(ch, ch) for ch in s)

print(to_lowercase("PYTHON123"))

```
